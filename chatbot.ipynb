```python
# ===============================
# 1️⃣ Imports
# ===============================

from youtube_transcript_api import YouTubeTranscriptApi
from langchain_text_splitters import RecursiveCharacterTextSplitter
from sentence_transformers import SentenceTransformer

from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import HuggingFaceEmbeddings


# ===============================
# 2️⃣ Fetch YouTube Transcript
# ===============================

video_id = "J5_-l7WIO_w"

ytt_api = YouTubeTranscriptApi()

try:
    transcript = ytt_api.fetch(video_id, languages=['hi'])
    all_text = " ".join([entry['text'] for entry in transcript.to_raw_data()])
except Exception as e:
    print("Error:", e)
    all_text = ""

print("Transcript length:", len(all_text))


# ===============================
# 3️⃣ Split Transcript
# ===============================

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50
)

chunks = text_splitter.split_text(all_text)

print("Total chunks:", len(chunks))


# ===============================
# 4️⃣ Embedding Model
# ===============================

embedding_model = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)


# ===============================
# 5️⃣ Create Vector Store
# ===============================

vector_store = FAISS.from_texts(
    texts=chunks,
    embedding=embedding_model
)

print("Vector store created")


# ===============================
# 6️⃣ Create Retriever
# ===============================

retriever = vector_store.as_retriever(
    search_type="similarity",
    search_kwargs={"k":3}
)


# ===============================
# 7️⃣ Retrieval Test
# ===============================

query = "टेक्स्ट स्प्लिट क्या होता है?"

docs = retriever.invoke(query)
print("\nTop retrieved chunks:\n")

for i, doc in enumerate(docs):
    print(f"Chunk {i+1}:")
    print(doc.page_content)
    print("-"*50)
```

```python
# ===============================
# 1️⃣ Imports
# ===============================

from youtube_transcript_api import YouTubeTranscriptApi
from langchain_text_splitters import RecursiveCharacterTextSplitter
from sentence_transformers import SentenceTransformer

from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import HuggingFaceEmbeddings

from langchain_core.prompts import PromptTemplate
from langchain_groq import ChatGroq


# ===============================
# 2️⃣ Fetch YouTube Transcript
# ===============================

video_id = "J5_-l7WIO_w"

ytt_api = YouTubeTranscriptApi()

try:
    transcript = ytt_api.fetch(video_id, languages=['hi'])
    all_text = " ".join([entry['text'] for entry in transcript.to_raw_data()])
except Exception as e:
    print("Error:", e)
    all_text = ""

print("Transcript length:", len(all_text))


# ===============================
# 3️⃣ Split Transcript
# ===============================

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50
)

chunks = text_splitter.split_text(all_text)

print("Total chunks:", len(chunks))


# ===============================
# 4️⃣ Embedding Model
# ===============================

embedding_model = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)


# ===============================
# 5️⃣ Create Vector Store
# ===============================

# --- Augmentation: add chunk metadata ---
augmented_chunks = []
for i, chunk in enumerate(chunks):
    # simple keyword extraction (first 5 words longer than 4 letters)
    keywords = [word for word in chunk.split() if len(word) > 4][:5]
    augmented_chunks.append({
        "text": chunk,
        "metadata": {
            "chunk_index": i+1,
            "video_id": video_id,
            "keywords": keywords
        }
    })

# Use augmented texts for vector store
texts = [c["text"] for c in augmented_chunks]
metadatas = [c["metadata"] for c in augmented_chunks]

vector_store = FAISS.from_texts(
    texts=texts,
    embedding=embedding_model,
    metadatas=metadatas
)

print("Vector store created")


# ===============================
# 6️⃣ Create Retriever
# ===============================

retriever = vector_store.as_retriever(
    search_type="similarity",
    search_kwargs={"k":3}
)


# ===============================
# 7️⃣ Retrieval Test
# ===============================

query = "टेक्स्ट स्प्लिट क्या होता है?"

docs = retriever.invoke(query)
print("\nTop retrieved chunks:\n")

for i, doc in enumerate(docs):
    print(f"Chunk {i+1}:")
    print(doc.page_content)
    print("-"*50)


# ===============================
# 8️⃣ Augmentation + PromptTemplate for Groq
# ===============================

# Augment retrieved chunks with chunk numbers
retrieved_augmented = [f"[Chunk {i+1}]\n{doc.page_content}" for i, doc in enumerate(docs)]
context = "\n\n".join(retrieved_augmented)

# --- Console the context ---
print("\n===== CONTEXT SENT TO LLM =====\n")
print(context)
print("\n===============================\n")

# PromptTemplate
prompt_template = """
You are a helpful assistant. Use the context from a YouTube video to answer the question.

Context:
{context}

Question:
{question}

Answer concisely:
"""

PROMPT = PromptTemplate(
    input_variables=["context", "question"],
    template=prompt_template
)

# Format final prompt
final_prompt = PROMPT.format(context=context, question=query)

# --- Console the final prompt ---
print("\n===== FINAL PROMPT =====\n")
print(final_prompt)
print("\n========================\n")

# Initialize Groq LLM
llm = ChatGroq(model="llama-3.1-8b-instant", api_key=api_key)

# Get answer
response = llm.invoke(final_prompt)

print("\nAnswer from Groq:\n")
print(response.content)
```
