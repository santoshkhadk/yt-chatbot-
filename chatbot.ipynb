```python
from youtube_transcript_api import YouTubeTranscriptApi
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.schema import Document
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS


# 1️⃣ Provide YouTube video ID
video_id = "J5_-l7WIO_w"


# 2️⃣ Fetch transcript
transcript = YouTubeTranscriptApi.get_transcript(video_id)


# 3️⃣ Convert transcript to plain text
text = " ".join([t["text"] for t in transcript])


# 4️⃣ Convert into LangChain document
doc = Document(page_content=text, metadata={"source": video_id})


# 5️⃣ Split document into chunks
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=100
)

docs = text_splitter.split_documents([doc])


# 6️⃣ Initialize Sentence Transformer embeddings
embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)


# 7️⃣ Create FAISS vector database
vectorstore = FAISS.from_documents(docs, embeddings)


# 8️⃣ Save index locally
vectorstore.save_local("vector_db")


print("✅ Document ingestion completed!")
```

```python
from youtube_transcript_api import YouTubeTranscriptApi
from langchain_text_splitters import RecursiveCharacterTextSplitter 
from langchain_core.documents import Document
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS

# 1️⃣ Provide YouTube video ID
video_id = "J5_-l7WIO_w"

# 2️⃣ Fetch transcript
transcript_list = YouTubeTranscriptApi.list_transcripts(video_id)
transcript = transcript_list.find_transcript(['en']).fetch()  # fetch actual transcript

# 3️⃣ Convert transcript to plain text
text = " ".join([t["text"] for t in transcript])

# 4️⃣ Convert into LangChain document
doc = Document(page_content=text, metadata={"source": video_id})

# 5️⃣ Split document into chunks
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=100
)

docs = text_splitter.split_documents([doc])

# 6️⃣ Initialize Sentence Transformer embeddings
embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

# 7️⃣ Create FAISS vector database
vectorstore = FAISS.from_documents(docs, embeddings)

# 8️⃣ Save index locally
vectorstore.save_local("vector_db")

print("✅ Document ingestion completed!")
```
